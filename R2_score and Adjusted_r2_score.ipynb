```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv('/content/case_1.csv')
df
```

```python
X = df[['YearsExperience']]
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
df1 = df.copy()
```

```python
df['Predictions'] = reg.predict(X)
```

```python
df
```

```python
from sklearn.metrics import r2_score
print(f'R2_score value : {r2_score(y,df["Predictions"])}')
```

```python
df1.shape[1]
```

```python
ar2s = 1 - ((1 - r2_score(y,df["Predictions"])) * (df1.shape[0] - 1)) / (df1.shape[0] - df1.shape[1] - 1)
print(f"Adjusted r2_score : {ar2s}")
```
